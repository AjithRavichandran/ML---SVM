```python
import pandas as pd
```

```python
dataset=pd.read_csv('50_Startups.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]
independent
```

```python
dependent=dataset[['Profit']]
dependent
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4,4,))
plt.plot(independent,dependent)
plt.xlabel('x',fontsize='20')
plt.ylabel('y',fontsize='20')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
input_train,input_test,output_train,output_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler     # Standardisation is must in SVM
sc=StandardScaler()                                  #Standardisation is used  if the Input of maximum value and minimum value is larger, the predirct will be worse, so we need to do Standardisation (-1 to 1) or (-2 to 2)
input_train=sc.fit_transform(input_train)
input_test=sc.transform(input_test)
```

```python
input_train
```

```python
input_test
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='rbf',C=10)             # it is in circle shape
regressor.fit(input_train,output_train)
```

```python
regressor.intercept_    #starting point
```

```python
regressor.n_support_    #total input row
```

```python
predict=regressor.predict(input_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(output_test,predict)
```

```python
r2
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='poly',C=500)            # its shape is parabola, 11.	If we want to increase the prediction go with C like C=0.01,0.1,1,10,100
regressor.fit(input_train,output_train)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(output_test,predict)
```

```python
r2
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='sigmoid',C=2000)            # its in S shape
regressor.fit(input_train,output_train)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(output_test,predict)
```

```python
r2
```
